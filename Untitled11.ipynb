```python
df = pd.read_csv("/content/jobs_act_parliament.csv")
```

```python
MODEL = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL, use_fast=True)
config = AutoConfig.from_pretrained(MODEL) # used for id to label name
```

```python
MAX_SEQ_LENGTH = 128

# Function to perform sentiment analysis on each text
def analyze_sentiment(text):
    # Tokenize the text with specified maximum length
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding="max_length", max_length=MAX_SEQ_LENGTH)

    # Pass the tokenized text through the model
    outputs = model(**inputs)

    # Get the predicted sentiment label
    predicted_class = torch.argmax(outputs.logits, dim=1).item()

    # Decode the sentiment label to the corresponding sentiment name
    sentiment_label = model.config.id2label[predicted_class]

    return sentiment_label
```

```python
df_text = df[["text"]]
```

```python
df_text.head()
```

```python
df_text['sentiment'] = df_text['text'].apply(analyze_sentiment)
```

```python
df_text.head()
```

```python
df_party = df[["text", "new_party_id"]]
```

```python
df_merge = pd.merge(df_text, df_party,  how = "left", on = "text")
```

```python
df_merge.head()
```

```python
pd.crosstab(df_merge.new_party_id, df_merge.sentiment, normalize = "index")\
    .round(4)*100
```
